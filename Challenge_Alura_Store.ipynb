# Importación de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

# 1. Análisis de facturación

```python
facturacion1=tienda1['Precio'].sum()
facturacion2=tienda2['Precio'].sum()
facturacion3=tienda3['Precio'].sum()
facturacion4=tienda4['Precio'].sum()
print(f'facturacion: {facturacion1}')
print(f'facturacion: {facturacion2}')
print(f'facturacion: {facturacion3}')
print(f'facturacion: {facturacion4}')
```

```python
total_fact=[]
total_fact.append(facturacion1)
total_fact.append(facturacion2)
total_fact.append(facturacion3)
total_fact.append(facturacion4)
print(total_fact)

tiendas=['t1','t2','t3','t4']

plt.bar (x=tiendas, height=total_fact)
plt.show()
```

Conclusión 1: En la Gráfica se observa que la tienda que mas facturó en el periodo enero 2020- marzo 2023 fue la tienda 1. La que menos vendio la tienda 4.



# 2. Ventas por categoría

```python
# Agrupar las ventas por la categría
tienda1.groupby(['Categoría del Producto'])[['Precio']].sum().sort_values('Precio',ascending=False)
```

```python
# Agrupar las ventas por la categría
tienda2.groupby(['Categoría del Producto'])[['Precio']].sum().sort_values('Precio',ascending=False)
```

```python
# Agrupar las ventas por la categría
tienda3.groupby(['Categoría del Producto'])[['Precio']].sum().sort_values('Precio',ascending=False)
```

```python
# Agrupar las ventas por la categría
tienda4.groupby(['Categoría del Producto'])[['Precio']].sum().sort_values('Precio',ascending=False)
```

Conslusión 2: En las 4 tiendas se obseva la misma clasificación de ventas siendo los Electrónicos la categoría más vendida y Los libros la menos facturada.

# 3. Calificación promedio de la tienda

```python
calificacion_t1 = tienda1['Calificación'].mean()
calificacion_t2 = tienda2['Calificación'].mean()
calificacion_t3 = tienda3['Calificación'].mean()
calificacion_t4 = tienda4['Calificación'].mean()
print(calificacion_t1)
print(calificacion_t2)
print(calificacion_t3)
print(calificacion_t4)
```

```python
prom_calificaciones=[]
prom_calificaciones.append(calificacion_t1)
prom_calificaciones.append(calificacion_t2)
prom_calificaciones.append(calificacion_t3)
prom_calificaciones.append(calificacion_t4)

print(prom_calificaciones)
```

```python
tiendas = ['t1','t2','t3','t4']
x = tiendas
y = prom_calificaciones

plt.scatter(x, y)
plt.show()
```

Conslusión 3: Mediante el gráfico scatter, se observa que la tienda 3 y la 2 tienen un mejor promedio de calificación, en escala la tienda 4 la sigue con -0.03 puntos por debajo y la tienda 1 tiene menor promedio por debajo de 3.98.
